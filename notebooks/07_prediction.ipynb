# 07 — Predict New Plant Readings
Load a saved model and scaler, input new sensor readings, and predict plant health status.

```python
import joblib
import os
import numpy as np
import pandas as pd
# Load scaler and model (choose RF or SVM)
scaler = joblib.load('../../models/scaler.pkl') if os.path.exists('../../models/scaler.pkl') else None
model = joblib.load('../../models/random_forest_model.pkl')  # or svm_model.pkl
# Input new readings
input_dict = {
    'Soil_Moisture': 0,
    'Ambient_Temperature': 0,
    'Soil_Temperature': 0,
    'Humidity': 0,
    'Light_Intensity': 0,
    'Soil_pH': 0,
    'Nitrogen_Level': 0,
    'Phosphorus_Level': 0,
    'Potassium_Level': 0,
    'Chlorophyll_Content': 0,
    'Electrochemical_Signal': 0
}
# Fill input_dict with real values
X_new = pd.DataFrame([input_dict])
if scaler is not None:
    X_new_scaled = scaler.transform(X_new)
else:
    X_new_scaled = X_new.values
pred = model.predict(X_new_scaled)[0]
label_map = {0: 'Healthy', 1: 'Moderate Stress', 2: 'High Stress'}
print('Predicted Health Status:', label_map[pred])
```
