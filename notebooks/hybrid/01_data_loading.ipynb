# 01 — Data Loading
Load the raw dataset and inspect its structure.

```python
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Load dataset
data_path = Path('../../data/processed_dataset.csv')
if not data_path.exists():
    data_path = Path('../data/processed_dataset.csv')

df = pd.read_csv(data_path)
print("Shape:", df.shape)
df.head()
```

```python
# Basic info
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Class distribution
target_col = next((c for c in ['Plant_Health_Status', 'Health_Status'] if c in df.columns), None)
if target_col is None:
    raise KeyError(f"No target label column found. Available columns: {list(df.columns)}")

print("\nClass Distribution:")
print(df[target_col].value_counts())
print("\nClass %:")
print(df[target_col].value_counts(normalize=True).mul(100).round(2))
```
